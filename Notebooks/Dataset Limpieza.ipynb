```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Data/TA_restaurants_curated.csv",encoding='utf8')
```

```python
data
```

```python
# Observo bien los datos de los que se forma mi dataset.
data.info()
```

```python
# Ver si hay alguna fila completamente formada por valores "NaN" y de ser así eliminarlas.
data1 = data.dropna(how = "all")
```

```python
data1
```

```python
# Cambiar "NaN" por guiones.
data2 = data1.fillna('-')
```

```python
data2
```

```python
# Observo que hay columnas que tienen el nombre formado por varias palabras y esto es un problema para luego trabajar con ellas.
# Por tanto, cambio los nombres de esas columnas para solventar este inconveniente.
data2.rename(columns={'Cuisine Style':'Cuisine_style','Price Range':'Price_range','Number of Reviews':'Reviews_number'}, inplace=True) 
```

```python
data2
```

```python
print(data2['Price_range'].value_counts())
```

```python
# Cambiar los tipos de precio que aparecen. Quitar los símbolos de dolares y cambiarlo por algo mas sencillo de entender.
data3 = data2.replace({"$": "cheap", "$$ - $$$": "medium", "$$$$": "high", "-": "not-inf"})
```

```python
data3
```

```python
# Verificar que el cambio se ha realizado correctamente.
print(data3['Price_range'].value_counts())
```

```python
# Decido quitar las columnas que no me sirven para mi posterior análisis.
restaurant_europe_data = data3.drop(['Unnamed: 0', 'URL_TA', 'ID_TA'], axis=1)
```

```python
restaurant_europe_data 
```
