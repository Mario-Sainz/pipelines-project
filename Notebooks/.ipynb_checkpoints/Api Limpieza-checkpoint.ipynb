```python
api_data1 = api_data.get("data")
```

```python
api_data1
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(api_data1)
```

```python
df
```

```python
df.info()
```

```python
df_1 = df[['name', 'timezone', 'location_string', 'ranking_category', 'rating', 'price_level', 'cuisine']]
```

```python
df_1
```

```python
print(df_1['price_level'].value_counts())
```

```python
df_2 = df_1.replace({"$": "cheap", "$$ - $$$": "medium"})
```

```python
df_2
```

```python
print(df_2['price_level'].value_counts())
```

```python
df_3 = df_2.rename(columns = {'timezone': 'City', 'name': 'Name', 'rating': 'Rating', 'cuisine': 'Cuisine_style' }, inplace = False)
```

```python
df_3
```

```python
restaurant_america_data = df_3.drop(['location_string', 'ranking_category', 'price_level'], axis=1)
```
